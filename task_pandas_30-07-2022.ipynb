```python
9. average weekly response time for each agent
```

```python
def con_time_sec(time_s):
    hh,mm,ss =time_s.split(':')
    return int(hh)*3600 + int(mm)*60 + int(ss)
```

```python
avg_response_sec=[]
for t in df['Average Response Time']:
    avg_response_sec.append(con_time_sec(t))
avg_response_sec
df['avg_response_sec'] = avg_response_sec
```

```python
df.head()
```

```python
df.dtypes['avg_response_sec']
```

```python
df.groupby(['Agent Name','week_of_year'])['avg_response_sec'].mean()
```
